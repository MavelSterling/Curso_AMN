# **Serie de Taylor**


**Una función en un punto a**


$f(x) = f(a) + f'(a)(x-a)
+ f''(a)\dfrac {(x-a)^2}{2!}
+ f ^{'''}(a)\dfrac {(x-a)^3}{3!}+\ldots+f^{(n)} (a)
\dfrac {(x-a)^n}{n!} + E_n$

#**Punto 1**

## Desarrollar la serie de Taylor de grado 2 para la función: $f(x)= Cos(xy)$




```python
import sympy as sp

# Definir variables simbólicas
x, y,h,k = sp.symbols('x y h k')

# Definir la función
f = sp.cos(x * y)

# Calcular la primera derivada
f1 = sp.diff(f, x)
print("Primera derivada con respecto a x: ", f1) #con respecto a x

# Calcular la segunda derivada
f2 = sp.diff(f, y)
print("primera derivada con respecto y: ", f2)

# Calcular la tercera derivada
f3 = sp.diff(f1, y)
print("segunda derivada cruzada dx|dy: ", f3)

# Calcular la cuarta derivada
f4 = sp.diff(f1, x)
print("segunda derivada con respecto a x: ", f4)

f5 = sp.diff(f2, y)
print("segunda derivada con respecto a y: ", f5)

serieTaylorSegundoOrden= f + f*h + f*k + (1/2)*(f4*(h**2)+2*(f3*h*k)+f5*(k**2))

valorDex = 2
valorDey = 3
desplazamientoEnx = 0.1
desplazamientoEny = 0.05

aproximacion = serieTaylorSegundoOrden.subs({x: valorDex, y: valorDey, h: desplazamientoEnx, k: desplazamientoEny})
print(aproximacion)
```

```python
import sympy as sp
import numpy as np
from tabulate import tabulate
import math as mt

# Definir la función
func_1 = sp.cos(x * y)

# Calcular la primera derivada
f1 = sp.diff(func_1, x)
print("Primera derivada con respecto a x: ", f1) #con respecto a x

# Calcular la segunda derivada
f2 = sp.diff(func_1, y)
print("primera derivada con respecto y: ", f2)

# Calcular la tercera derivada
f3 = sp.diff(f1, y)
print("segunda derivada cruzada dx|dy: ", f3)

# Calcular la cuarta derivada
f4 = sp.diff(f1, x)
print("segunda derivada con respecto a x: ", f4)

f4_y = sp.diff(f2, y)
print("segunda derivada con respecto a y: ", f4_y)


a_i = np.array([0.1,0.05])        #Vector a
x_i = np.array([2,3]) #[x,y]

#Valor final de x = x_(i+1)
h = np.subtract(x_i, a_i )    #Incremento
n = 5           #Orden del polinomio

#Funcion y sus derivadas
def function_f1(x,y):
  F=[]                          #Vector de la funcion y sus derivadas
  F.append(func_1)          #f(x)
  F.append(f1)  #f'(x) Primera derivada con respecto a x
  F.append(f2)     #f'(x) Primera derivada con respecto a y
  F.append(f4)      #f''(x) segunda derivada con respecto a x
  F.append(f4_y)   #f''(x)segunda derivada con respecto a y
  F.append(f3)   #f''(x)  segunda derivada cruzada dx|dy
  return (F)



def Taylor_Serie1(xi, h, n):
  f_i = function_f1(x_i[0],x_i[1]) #Vector que contendra la función original y sus derivadas| La función f esta siendo evaluada desde 0 con xi
  f_i11 = function_f1(x_i[0],x_i[1])[0] # Valor real que debe dar la función evaluada en ese punto.
  T= [] #Terminos de la serie
  Orden= [] # Numeración del orden

  tablaTylor = []
  sumatoriaFunc= 0

  for i in range(n+1):
    #Calcular la serie con la función y sus derivadas
    T.append(f_i[i]* (h**i)/np.math.factorial(i)) # La función f_i[i] recorre todas las derivadas que guardo inicialmente, iniciando desde la primera y calcula la serie

    Orden.append(i) # Orden n

    tablaTylor.append([Orden[i], T[i]])

  print( tabulate(tablaTylor, headers = ["Orden Derivada", "f(x(i))"]))

#Sumatoria y resultado de la serie
  for termino in T:
    sumatoriaFunc+=termino
  print(f"\n Cos(x*y) con serie de Taylor y evaluado en ({a_i[0]},{a_i[0]}) es aproximadamente: {(sumatoriaFunc)}")

#Taylor_Serie()
table = Taylor_Serie1 (a_i, h, n)
print(table)
```

```python
import sympy as sp
import numpy as np
from tabulate import tabulate
import math as mt

# Definir variables simbólicas
x, y = sp.symbols('x y')

# Definir la función
f = sp.cos(x + y)

# Calcular la primera derivada
f1 = sp.diff(f, x)
print("Primera derivada con respecto a x: ", f1) #con respecto a x

# Calcular la segunda derivada
f2 = sp.diff(f, y)
print("primera derivada con respecto y: ", f2)

# Calcular la tercera derivada
f3 = sp.diff(f1, y)
print("segunda derivada cruzada dx|dy: ", f3)

# Calcular la cuarta derivada
f4 = sp.diff(f1, x)
print("segunda derivada con respecto a x: ", f4)

f4 = sp.diff(f2, y)
print("segunda derivada con respecto a y: ", f4)

print("\n")

a_i = np.array([0,0])        #Vector a
x_i = np.array([0,0]) #[x,y]
#Valor final de x = x_(i+1)
h = np.subtract(x_i, a_i )    #Incremento
n = 5           #Orden del polinomio

#Funcion y sus derivadas
def f(x,y):
  F=[]                          #Vector de la funcion y sus derivadas
  F.append(mt.cos(x+y))          #f(x)
  F.append(-(mt.sin(x + y)))  #f'(x) Primera derivada con respecto a x
  F.append(-(mt.sin(x + y)))     #f'(x) Primera derivada con respecto a y
  F.append(-(mt.cos(x + y)))      #f''(x) segunda derivada con respecto a x
  F.append(-(mt.cos(x + y)))   #f''(x)segunda derivada con respecto a y
  F.append(-(mt.cos(x + y)))   #f''(x)  segunda derivada cruzada dx|dy
  return (F)



def Tylor(xi, h, n):
  f_i = f(x_i[0],x_i[1]) #Vector que contendra la función original y sus derivadas| La función f esta siendo evaluada desde 0 con xi
  f_i11 = f(x_i[0],x_i[1])[0] # Valor real que debe dar la función evaluada en ese punto.
  T= [] #Terminos de la serie
  Orden= [] # Numeración del orden

  tablaTylor = []
  sumatoriaFunc= 0

  for i in range(n+1):
    #Calcular la serie con la función y sus derivadas
    T.append(f_i[i]* (h**i)/np.math.factorial(i)) # La función f_i[i] recorre todas las derivadas que guardo inicialmente, iniciando desde la primera y calcula la serie

    Orden.append(i) # Orden n

    tablaTylor.append([Orden[i], T[i]])

  print( tabulate(tablaTylor, headers = ["Orden Derivada", "f(x(i))"]))

#Sumatoria y resultado de la serie
  for termino in T:
    sumatoriaFunc+=termino
  print(f"\n Cos(x+y) con serie de Tylor y evaluado en ({a_i[0]},{a_i[1]}) es aproximadamente: {(sumatoriaFunc)}")

#Tylor()
table = Tylor (a_i, h, n)
print(table)
```
