## Método Jacobi

Algortimo:
```
input n, (a_{ij}), (b_{i}), (x_{i}), M
for k = 1 to M do
        for i =1 to n do
           u_{i} = (b_{i} - \sum_{j=1}^{n} a_{ij} x_{j}) /a_{ii} \\
        end do \\
        for i =1 to n do
            x_{i} = u_{i}
        end do
        output k, (x_{i})  
end do
```

````python
def metodo_jacobi(n, a, b, x, M):
    """
    Método de Jacobi para resolver sistemas de ecuaciones lineales.

    Parámetros:
    - n: número de variables en el sistema de ecuaciones
    - a: matriz de coeficientes de tamaño nxn (una lista de listas)
    - b: vector de términos independientes de tamaño n
    - x: vector de valores iniciales de las variables de tamaño n
    - M: número máximo de iteraciones

    Retorna:
    - iteraciones: lista de tuplas (k, x) con las iteraciones del algoritmo
    """
    iteraciones = []
    for k in range(1, M+1):
        u = [0] * n  # Vector temporal para almacenar los nuevos valores de x

        # Calcula los nuevos valores de x
        for i in range(n):
            sumatoria = 0
            for j in range(n):
                if j != i:
                    sumatoria += a[i][j] * x[j]
            u[i] = (b[i] - sumatoria) / a[i][i]

        # Actualiza los valores de x con los nuevos valores obtenidos
        for i in range(n):
            x[i] = round(u[i],4)

        iteraciones.append((k, x.copy()))
        print(f"Iteración {k}:")
        print(f"Valores aprox. de las variables: {x}")
        print(" ")

    return iteraciones
````

````python
# Datos del problema a

n = 3
a = [[4, 1, -1],
     [-1, 3, 1],
     [2, 2, 5]]
b = [5, -4, 1]
x = [0, 0, 0]
M = 10

# Iterar utilizando el método Jacobi y mostrar los valores de cada iteración
iteraciones = metodo_jacobi(n, a, b, x, M)
````

## Método Gauss-Seidel

Algortimo:
```
input n, (a_{ij}), (b_{i}), (x_{i}), M
for k = 1 to M do
        for i =1 to n do
              x_{i} = (b_{i} - \sum_{j=1}^{n} a_{ij} x_{j}) /a_{ij} \\
        end do
output k_{i} , (x_{i}))
end do
```

 Algoritmo de Gauss-Seidel para resolver sistemas de ecuaciones lineales.

    Parámetros:
    - n: número de variables en el sistema de ecuaciones
    - a: matriz de coeficientes (una lista de listas)
    - b: vector de términos independientes
    - x: vector de valores iniciales de las variables
    - M: número máximo de iteraciones

    Retorna:
    - k: número de iteraciones realizadas
    - x: vector de valores aproximados de las variables


````python
def gauss_seidel_iteration(n, a, b, x, M):
    iteraciones = []
    for k in range(1, M+1):
        for i in range(n):
            sum_term = 0
            for j in range(n):
                if j != i:
                    sum_term += a[i][j] * x[j]
            x[i] = (b[i] - sum_term) / a[i][i]

        print(f"Iteración {k}: {x}")
        iteraciones.append((k, x.copy()))

    return iteraciones
````

## Método Richarson

Algoritmo:



```
input n, (a_{ij}), (b_{i}), (x_{i}), M
for k = 1 to M do
        for i =1 to n do
              r_{i} <- (b_{i} - \sum_{j=1}^{n} a_{ij} x_{j})
        end do
        for i= 1 to n do
                x_{i} <- x_{i}. + r_{i}
        end do
end do
output k , (x_{i}), r_{i}
```



````python
def metodo_richardson(n, a, b, x, M):
    """
    Método Richardson para resolver sistemas de ecuaciones lineales y calcular los residuos.

    Parámetros:
    - n: número de variables en el sistema de ecuaciones
    - a: matriz de coeficientes de tamaño nxn (una lista de listas)
    - b: vector de términos independientes de tamaño n
    - x: vector de valores iniciales de las variables de tamaño n
    - M: número máximo de iteraciones

    Retorna:
    - iteraciones: lista de tuplas (k, x) con las iteraciones del algoritmo
    """
    iteraciones = []
    for k in range(M+1):
        r = [0] * n  # Vector de residuos
        for i in range(n):
            for j in range(n):
                r[i] += a[i][j] * x[j]  # Cálculo de los residuos
            r[i] = b[i] - r[i]  # Actualización del residuo
        for i in range(n):
            x[i] += r[i]  # Actualización de la solución

        iteraciones.append((k, x.copy()))
        print(f"Iteración {k}: {[round(val, 4) for val in x]}")


    return iteraciones

````

## Método Gradicente descendiente

Algortimo:


```
input x, A, b, M
output 0, x
for k = 1 to M do
    v = b - Ax
    t = \langle v, v \langle / \langle v, Av \langle  # producto punto
    x = x + tv
output k, x
end do
```





````python
def gradient_descent(x, A, b, M):
    """
    Método del gradiente descendiente para resolver sistemas de ecuaciones lineales.

    Parámetros:
    - x: vector inicial de tamaño n
    - A: matriz de coeficientes de tamaño nxn (una lista de listas)
    - b: vector de términos independientes de tamaño n
    - M: número máximo de iteraciones

    Retorna:
    - iteraciones: lista de tuplas (k, x) con las iteraciones del algoritmo
    """
    iteraciones = []
    for k in range(M):
        v = [b[i] - sum(A[i][j] * x[j]
                        for j in range(len(x)))
                            for i in range(len(x))]
        t = sum(v[i] * v[i]
                for i in range(len(x))) / sum(v[i] * sum(A[i][j] * v[j]
                                                                     for j in range(len(x)))
                                                                            for i in range(len(x)))
        x = [round(x[i] + t * v[i], 4)
                for i in range(len(x))]
        iteraciones.append((k + 1, x.copy()))

        print(f"Iteración {k+1}:")
        print(f"Valores aproximados de las variables: {[round(val, 4) for val in x]}")
        print()

    return iteraciones
````

````python
# Datos del problema a
A = [[4, 1, -1],
     [-1, 3, 1],
     [2, 2, 5]]
b = [5, -4, 1]
x = [0, 0, 0]
M = 4

# Iterar utilizando el método gradiente descendiente y mostrar los valores de cada iteración
iteraciones = gradient_descent(x, A, b, M)
````

## Método SOR

Donde w es un factor con los siguientes rangos de valores:

*  Sub-relajación 0 < w < 1 se logra la convergencia en sistemas divergentes.
*  Sobre-relajación 1 < w < 2 acelera la convergencia de sistemas convergentes.



````python
def metodo_sor(n, a, b, x, w, M):
    """
    Método SOR (Successive Over-Relaxation) para resolver sistemas de ecuaciones lineales.

    Parámetros:
    - n: número de variables en el sistema de ecuaciones
    - a: matriz de coeficientes de tamaño nxn (una lista de listas)
    - b: vector de términos independientes de tamaño n
    - x: vector de valores iniciales de las variables de tamaño n
    - w: factor de relajación (0 < w < 2)
    - M: número máximo de iteraciones

    Retorna:
    - iteraciones: lista de tuplas (k, x) con las iteraciones del algoritmo
    """
    iteraciones = []
    for k in range(1, M+1):
        for i in range(n):
            sumatoria1 = sum(a[i][j] * x[j] for j in range(i))
            sumatoria2 = sum(a[i][j] * x[j] for j in range(i+1, n))
            x[i] = (1 - w) * x[i] + (w / a[i][i]) * (b[i] - sumatoria1 - sumatoria2)

        iteraciones.append((k, x.copy()))

        print(f"Iteración {k}:")
        print(f"Valores aproximados de las variables: {x}")
        print()

    return iteraciones
````

````python
# Datos del problema a
n = 3
a = [[4, 1, -1],
     [-1, 3, 1],
     [2, 2, 5]]
b = [5, -4, 1]
x = [0, 0, 0]
w = 1.2
M = 3

# Iterar utilizando el método SOR y mostrar los valores de cada iteración
iteraciones = metodo_sor(n, a, b, x, w, M)
````

````python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

%matplotlib inline

def calcularError(x, xAnt):
    n = len(x)
    error = max(abs(x[i] - xAnt[i]) for i in range(n))
    return error

def metodoJacobiGrafica(n, a, b, x, M):
    iteraciones = metodo_jacobi(n, a, b, x, M)

    fig, ax = plt.subplots()
    line, = ax.plot([], [], 'b-')

    ax.set_xlim(0, M)
    ax.set_ylim(0, max(calcularError(x, iteracion[1]) for iteracion in iteraciones))

    iteracionesEjex = []
    iteracionesEjeY = []

    def actualizarGrafico(iteracion):
        xActual = x.copy()
        error = calcularError(xActual, iteracion[1])

        iteracionesEjex.append(iteracion[0])
        iteracionesEjeY.append(error)

        line.set_data(iteracionesEjex, iteracionesEjeY)
        ax.relim()
        ax.autoscale_view()
        return line,

    anim = FuncAnimation(fig, actualizarGrafico, frames=iteraciones, blit=True)
    plt.close()  # Close the initial figure

    return HTML(anim.to_html5_video())

````

````python
##Sor con grafica en vivo
import matplotlib.pyplot as plt
from IPython.display import display, clear_output


def calcularError(x, xAnt):
    n = len(x)
    error = max(abs(x[i] - xAnt[i]) for i in range(n))
    return error

def metodoSorGrafica(n, a, b, x, w, M):

    iteraciones = metodo_sor(n, a, b, x, w, M)

    fig, ax = plt.subplots()
    line, = ax.plot([], [], 'b-')

    ax.set_xlim(0, M)
    ax.set_ylim(0, max(calcularError(x, iteracion[1]) for iteracion in iteraciones))

    iteracionesEjex = []
    iteracionesEjeY = []

    def actualizarGrafico(iteracion):
        xActual = x.copy()
        error = calcularError(xActual, iteracion[1])

        iteracionesEjex.append(iteracion[0])
        iteracionesEjeY.append(error)

        line.set_data(iteracionesEjex, iteracionesEjeY)
        ax.relim()
        ax.autoscale_view()
        clear_output(wait=True)
        display(fig)
        print(f"Iteración {iteracion[0]}:")
        print(f"Valores aproximados de las variables: {iteracion[1]}")
        plt.pause(0.1)

    todas_las_iteraciones = []
    for iteracion in iteraciones:
        todas_las_iteraciones.append(iteracion)
        actualizarGrafico(iteracion)

    plt.close(fig)

    return todas_las_iteraciones

````

````python
# Se revisa si las matrices de los ejecicios son diagonal dominante

def diagonal_dominante(A):
    num_filas = len(A)  # Obtener el número de filas
    num_columnas = len(A[0])  # Obtener el número de columnas

    for i in range(num_filas):  # Recorrer cada fila
        suma = 0  # Inicializar la suma de los elementos de la fila

        for j in range(num_columnas):  # Recorrer cada columna de la fila actual
            if j != i:  # Excluir la diagonal principal
                suma += abs(A[i][j])  # Sumar el valor absoluto del elemento a la suma

        if abs(A[i][i]) <= suma:  # Comparar el valor absoluto del elemento diagonal con la suma
            return False  # Si no se cumple la condición de diagonal dominante, retornar False

    # Si se completa el recorrido sin romper la condición, la matriz es diagonal dominante y se retorna True
    return True
````

````python
# Datos del problema a
a = [[4, 1, -1],
     [-1, 3, 1],
     [2, 2, 5]]

print("Es diagonal dominante:", diagonal_dominante(a))
````

````python
# Datos del problema b
a = [[-2, 1, 1/2],
     [1, -2, -1/2],
     [0, 1, 2]]
print("Es diagonal dominante:", diagonal_dominante(a))
````

````python
# Datos del problema c
a = [[4, 1, -1, 1],
     [1, 4, -1, -1],
     [-1, -1, 5, 1],
     [1, -1, 1, 3]]
print("Es diagonal dominante:", diagonal_dominante(a))
````

````python
# Datos del problema d
a = [[4, -1, 0, -1, 0, 0],
     [-1, 4, -1, 0, -1, 0],
     [0, -1, 4, 0, 0, -1],
     [-1, 0, 0, 4, -1, 0],
     [0, -1, 0, -1, 4, -1],
     [0, 0, -1, 0, -1, 4]]
print("Es diagonal dominante:", diagonal_dominante(a))
````



*   Se revisaron si las matrices de los ejecicios y se encontró que solo la matriz del punto c no es matriz diagonalmente dominante, por lo que los resultados con esa matriz no son óptimos para algunos de estos métodos como los son Jacobi y Gauss-Seidel, ya que ellos utilizan matrices diagonalmente dominante.

*   Los métodos de Jacobi y Gauss-Seidel son iterativos, sin embargo el Gauss-Seidel converge más rápido que el método de Jacobi, ya que va actualizando los valores aproximados con cada iteración.

*   El método SOR no necesariamente requiere que la matriz sea diagonalmente dominante. Sin embargo, un factor de relación que tiene unos intervalos:

1.   Sub-relajación 0 < w < 1 se logra la convergencia en sistemas divergentes.
2.   Sobre-relajación 1 < w < 2 acelera la convergencia de sistemas convergentes.

En el caso que el factor de relajación no esta en este intervalo, la convergencia es más lenta. Para los ejercicios se utilizó un w = 1.5, por lo que se encuentra en el intervalo de sobre-relajación.
      
*   El método de Gradiente descendiente su convergencia se da en matrices simétricas y definidas positivas. De lo contrario, su convergencia es muy lenta o puede que no converga.

*   El método de richardson converge en matrices donde su radio espectral es menor que 1, en caso contrario este método diverge. El radio espectral es el valor absoluto del valor propio más grande de la matriz.
